```python
import numpy as np
rng = np.random.default_rng(seed= 1701) # seed for reproducibility

x1= rng.integers(10, size= 6) 
print("x1 : ",x1)
x2= rng.integers(10, size=(3,4)) # two-dimensional array
print("x2 : ",x2)
x3= rng.integers(10, size= (3, 4, 5)) #three-dimensional array
print("x3 : ", x3)
```

```python
print("x3 ndim: ", x3.ndim)
print("x3 shape: ", x3.shape)
print("x3 size: ", x3.size)
print("dtype: ", x3.dtype)
```

## Subarrays as No-Copy Views

```python
print(x2)
```

```python
x2_sub = x2[:2, :2]
print(x2_sub)
```

```python
x2_sub[0,0] = 99
print(x2_sub)
```

```python
print(x2)
```

```python
x2_sub_copy = x2[:2, :2].copy()
print(x2_sub_copy)
```

```python
# if we modify this subarray, the original array is not touched:
x2_sub_copy[0, 0]= 42
print(x2_sub_copy)
```

```python
print(x2)
```

## Array Concatenation and Splitting

### Concatenation of Arrays

```python
x = np.array([1, 2, 3])
y = np.array([3, 2, 1])
np.concatenate([x, y])
```

```python
z = np.array([99, 99, 99])
print(np.concatenate([x, y, z]))
```

```python
# can also be used for 2D arrays:
grid = np.array([[1, 2, 3], 
                 [4, 5, 6]])
```

```python
# concatenate along the first axis
np.concatenate([grid, grid])
```

```python
# concatenate along the second axis (zero-indexed)
np.concatenate([grid, grid], axis= 1)
```

```python
# vertically stack the arrays
np.vstack([x, grid])
```

```python
# horizontal stack the arrays
y = np.array([[99], [99]])
np.hstack([grid, y])
```

### Splitting of Arrays

```python
x = [1, 2, 3, 99, 99, 3, 2, 1]
x1, x2, x3= np.split(x, [3, 5]) # split the array at index 3 and index 5
print(x1, x2, x3)

# [1 2 3 | 99 99 | 3 2 1]
#        ↑       ↑
#      index3  index5
```

```python
grid = np.arange(16).reshape((4, 4))
grid
```

```python
upper, lower = np.vsplit(grid, [2])
print(upper)
print(lower)
```

```python
left, right = np.hsplit(grid, [2])
print(left)
print(right)
```
